```python
import pandas as pd
vault = pd.read_csv('vaultbrewing.csv')
vault.fillna(0)
vault.head()
```

```python
import matplotlib.pyplot as plt
vault['Review_Date'] = pd.to_datetime(vault['Review_Date'])
vault = vault.sort_values(by = 'Review_Date')
vault.head(5)
```

```python
vault.describe()
```

```python
plt.figure(figsize = (20,10))
plt.plot(vault.Review_Date,vault.Rating,'bo',)
plt.title('Reviews of Vault Brewing Co',fontsize = 30,fontweight = 'bold')
plt.xlabel('Date of Review', fontsize = 15)
plt.ylabel('Number of Stars Rated on Yelp', fontsize = 15)
plt.yticks(fontsize = 12)
plt.xticks(fontsize =12)
```

```python
v = vault.copy()
v.index = v.Review_Date
plt.figure(figsize = (15,10))
plt.plot(v.resample('M').median(),'r-*')
plt.title('Median Yelp Review by Month',fontsize = 20, fontweight = 'bold')
plt.ylabel('Number of Stars Given',fontsize = 15)
plt.xlabel('Time Period Review Posted',fontsize = 15)
```

```python
plt.figure(figsize = (15,10))
plt.plot(v.resample('M').count(),'*-')
plt.title('Number of Yelp Reviews Grouped by Month',fontsize = 20,fontweight = 'bold')
plt.xlabel('Time Period Review Posted',fontsize = 15)
plt.ylabel('Number of Reviews Posted',fontsize = 15)
```

```python
import numpy as np
plt.figure(figsize = (15,10),)
x = pd.DataFrame(v.Rating.resample('M').count())
y = pd.DataFrame(v.Rating.resample('M').mean())
s = np.array(x.Rating)
s[np.isnan(s)]=0
s = np.delete(s, np.argwhere(s == 0))
t = np.array(y.Rating)
t[np.isnan(t)]=0
t = np.delete(t, np.argwhere(t == 0))

def best_fit(s,t):
    w, b  = np.polyfit(s,t,deg=1)
    line  = w * s + b
    return line

line = best_fit(s,t)

plt.plot(s,t,'k*')
plt.plot(s,line,'r--')

plt.title('Review Count per Month vs Average Yelp Rating',fontsize = 20, fontweight = 'bold')
plt.xlabel('Number of Reviews Posted in Month',fontsize = 15)
plt.ylabel('Average Yelp Review Rating in Month',fontsize = 15)
```

```python
line = best_fit(s,t)
plt.figure(figsize = (15,10))
plt.plot(s,t,'k*')
plt.plot(s,line,'r--')

plt.title('Review Count per Month vs Average Yelp Rating',fontsize = 20, fontweight = 'bold')
plt.xlabel('Number of Reviews Posted in Month',fontsize = 15)
plt.ylabel('Average Yelp Review Rating in Month',fontsize = 15)
```

```python
#Removing an outlier from the previous plot, where they only received one review that month that was a one-star review.

q = s
q = np.delete(q,10)

r = t
r = np.delete(r,10)
```

```python
line = best_fit(q,r)
plt.figure(figsize = (15,10))
plt.plot(q,r,'k*')
plt.plot(q,line,'r--')

plt.suptitle('Removed one month skewing the dataset',fontsize =18)
plt.title('Review Count per Month vs Average Yelp Rating',fontsize = 20, fontweight = 'bold')
plt.xlabel('Number of Reviews Posted in Month',fontsize = 15)
plt.ylabel('Average Yelp Review Rating in Month',fontsize = 15)
```
